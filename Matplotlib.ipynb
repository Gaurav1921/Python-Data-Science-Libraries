## Load Necessary Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## Basic Graph

```python
# defifning values for x and y
x = [0,1,2,3,4,5]
y = [0,2,4,6,8,10]

# to resize your graph & to specify how many pixels
plt.figure(figsize=(10,3), dpi=100)

####################################################### Line No 1  #########################################################

# Keyword Argument Notation
# to change the color and width of the line adn add the marker(dots) to the line
plt.plot(x,y, label='2x', color='blue', linewidth=2, marker='.', markersize=10, linestyle='--')
plt.legend()

# Shorthand Notation
# fmt = '[color][marker][line]'
# plt.plot(x,y,'b^--', label='2x')

####################################################### Line No 2  #########################################################

# Using numpy in matplotlib

# define x and y
x2 = np.arange(0,4,0.5)

# Plot part of the graph as line
plt.plot(x2[:6], x2[:6]**2,'r', label='x^2')

# Plot remainder of dot as a graph
plt.plot(x2[5:], x2[5:]**2,'r--', label='x^2')

# to give title, fontname/size
plt.title('Our 1st Graph', fontdict={'fontname' : 'Comic Sans MS', 'fontsize' : 20})

# to print label X & Y axes
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

# to print until how many number on number line we want for X & Y axes (scale of my graph)
plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8,9,10])

# Show plot. Even if not written, it's okay
plt.show()    

# to save a graph
plt.savefig('mygraph.pdf')
```

## Bar Charts

```python
labels = ['A', 'B', 'C']
values = [1,4,2]

bars = plt.bar(labels, values)

bars[0].set_hatch('/')
bars[1].set_hatch('*')
bars[2].set_hatch('.')

# or use this notation
"""
patterns = ['/','*','.']
for bar in bars:
    bar.set_hatch(patterns.pop(0))
"""

plt.figure(figsize=(6,4))

plt.show()
```
